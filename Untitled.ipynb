```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
url = "https://raw.githubusercontent.com/umaimehm/Intro_to_AI_2021/main/assignment2-machinelearning/Ruter-data.csv"
df_original = pd.read_csv(url, sep=";") 
df_original.head()
```

#### Filter the original dataframe with only the columns 'Dato', 'Linjenavn' and 'Passasjerer_Ombord'

```python
df = df_original.filter(['Dato','Linjenavn','Passasjerer_Ombord'], axis=1)
```

#### Replace all negative values of 'Passasjerer_Ombord' column with a median value.

```python
median_passasjerer_ombord = df['Passasjerer_Ombord'].median()
for index, row in df.iterrows():
    if row['Passasjerer_Ombord'] < 0:
        df.loc[index,'Passasjerer_Ombord'] =  median_passasjerer_ombord
```

#### Convert 'Dato' column to *date* then to *numeric*.

```python
df["Dato"] = pd.to_datetime(df["Dato"])
df["Dato"] = pd.to_numeric(df["Dato"])
```

#### Find all unique bus lines.

```python
linjer = df["Linjenavn"].unique()
```

#### Create a df for each unique busline.

```python
array = []
for key, value in enumerate(linjer):
    array.append(df[df['Linjenavn'] == value])
```

#### Sum the number of passenger per date in each df.

```python
array1 = []
for key, value in enumerate(array):
    array1.append(value.groupby(['Dato', 'Linjenavn'], as_index=False).agg({'Passasjerer_Ombord':'sum'}))
```

#### Create a scatter plot for each df.

```python
for key, value in enumerate(array1):
    t = value['Linjenavn'].unique()[0]
    value.plot.scatter(x="Dato", y="Passasjerer_Ombord", title='Linje: ' + t)
```
